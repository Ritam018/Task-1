Defining the values of x and y for training

```python

x=d.drop('SalePrice',axis=1)
y=d['SalePrice']
```

Fitting the training data into the Linear Regression model

```python
lr=LinearRegression()
lr.fit(x,y)
```

```python
d['SalePrice'].head()
```

Predicting the values of output variable

```python

y_pred=lr.predict(d.iloc[:10,1:])
y_pred
```

Printing the model accuracy for the training data

```python
lr.score(x,y)
```
